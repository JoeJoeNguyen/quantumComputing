```python
from qiskit import *
import matplotlib as mpl 
```

```python
qc1 = QuantumCircuit(2,2)
qc1.h(0)
qc1.cx(0,1)
qc1.draw('mpl')
```

draw the operator

```python
from qiskit.quantum_info import Operator


operator_qc1 = Operator(qc1)

operator_qc1.draw('latex')
```

```python
import numpy as np
from qiskit.circuit import Parameter    

parameter1 = np.pi
parameter2 = Parameter('θ') 
qc1.rx(parameter1, 0)
qc1.rx(parameter2, 1)
qc1.draw('mpl')
```

```python
qc1 = qc1.assign_parameters({parameter2: 2.1})
qc1.draw('mpl')
```

```python
operator_qc1 = Operator(qc1)

operator_qc1.draw('latex')
```

```python
qc1.measure([0,1],[0,1])
qc1.draw('mpl')
```

Measurement counts

```python
from qiskit.visualization import plot_histogram
from qiskit_aer import Aer
from qiskit import transpile
shots = 1000

simulator = Aer.get_backend('qasm_simulator')

qc1 = transpile(qc1, simulator)

result = simulator.run(qc1, shots= shots).result()

counts = result.get_counts(qc1)

plot_histogram(counts)
```
