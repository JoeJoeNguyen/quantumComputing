$\text{Rx, Ry, Rz Rotation Gates}$

$\text{Rx gate}$

```python
#importing

import qiskit
import numpy as np

from qiskit import QuantumCircuit,QuantumRegister,transpile

from qiskit_aer import AerSimulator, Aer
from qiskit.visualization import plot_histogram, plot_bloch_vector, plot_bloch_multivector
from qiskit.quantum_info import Statevector
import math
from math import pi
```

```python
# Rx Gate

theta_x = pi/8

qc_rx = QuantumCircuit(1,1, name='qc')

qc_rx.rx(theta_x,0)

qc_rx.draw('mpl')
```

$\text{Bloch Sphere}$

```python
backend = Aer.get_backend('statevector_simulator')

result = backend.run(qc_rx).result().get_statevector(qc_rx, decimals=3)

print("The quantum State is: ",result)

plot_bloch_multivector(result)
```

$\text{Unitary of Rx gate}$

```python
backend = Aer.get_backend('unitary_simulator')

qc_rx_unitary = backend.run(qc_rx).result().get_unitary(qc_rx, decimals=3)

print("Circuit unitary:\n", np.asarray(qc_rx_unitary).round(5))
```

$\text{Ry gate}$

```python
#Ry Gate

theta_y = -pi/8

qc_ry = QuantumCircuit(1,1, name='qc')

qc_ry.ry(theta_y,0)

qc_ry.draw('mpl')
```

$\text{Bloch Sphere}$

```python
backend = Aer.get_backend('statevector_simulator')

result = backend.run(qc_ry).result().get_statevector(qc_ry, decimals=3)

print("The quantum State is: ",result)

plot_bloch_multivector(result)
```

$\text{Unitary Matrix for Ry gate}$

```python
backend = Aer.get_backend('unitary_simulator')

qc_ry_unitary = backend.run(qc_ry).result().get_unitary(qc_ry, decimals=3)

print("Circuit unitary:\n", np.asarray(qc_ry_unitary).round(5))
```

$\text{Rz Gate}$

```python
#Rz Gate

theta_z = 2*pi

qc_rz = QuantumCircuit(1,1, name='qc')

qc_rz.x(0)

qc_rz.rz(theta_z,0)

qc_rz.draw('mpl')
```

$\text{Bloch Sphere}$

```python
backend = Aer.get_backend('statevector_simulator')

result = backend.run(qc_rz).result().get_statevector(qc_rz, decimals=3)

print("The quantum State is: ",result)

plot_bloch_multivector(result)
```

$\text{Unitary Matrix of Rz gate}$

```python
backend = Aer.get_backend('unitary_simulator')

qc_rz_unitary = backend.run(qc_rz).result().get_unitary(qc_rz, decimals=3)

print("Circuit unitary:\n", np.asarray(qc_rz_unitary).round(5))
```
