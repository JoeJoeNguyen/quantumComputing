$S, S^\dagger \text{ and } T, T^\dagger \text{ gate}$

```python
#importing

import qiskit
import numpy as np

from qiskit import QuantumCircuit,QuantumRegister,transpile

from qiskit_aer import AerSimulator, Aer
from qiskit.visualization import plot_histogram, plot_bloch_vector, plot_bloch_multivector
from qiskit.quantum_info import Statevector


```

$\text{S gate}$

```python
qc_s = QuantumCircuit(1,1, name="qc")
qc_s.x(0)
qc_s.s(0)

qc_s.draw('mpl')
```

$\text{Bloch Sphere}$

```python
#mapping the qubit on the bloch sphere
qc_s = QuantumCircuit(1,1, name="qc")
qc_s.x(0)
qc_s.s(0)
qc_s.save_statevector()
backend = AerSimulator(method="statevector")

qc_s = transpile(qc_s,backend)

result = backend.run(qc_s).result()

Statevector = result.get_statevector(qc_s, decimals=3)

print ("the quantum state is: ", Statevector)

plot_bloch_multivector(Statevector)
```

$\text{unitary matrix S gate}$

```python
#unitary matrix
qc_s = QuantumCircuit(1,1, name="qc")
qc_s.x(0)
qc_s.s(0)
backend = Aer.get_backend("unitary_simulator")

s_unitary = backend.run(qc_s)

s_unitary.result().get_unitary(qc_s)
```

$S^\dagger \text{ gate}$ 

```python
qc_sdg = QuantumCircuit(1,1, name="qc")
qc_sdg.x(0)
qc_sdg.sdg(0)

qc_sdg.draw('mpl')
```

$\text{Bloch Sphere}$

```python
#mapping the qubit on the bloch sphere
qc_sdg = QuantumCircuit(1,1, name="qc")
qc_sdg.x(0)
qc_sdg.sdg(0)
qc_sdg.save_statevector()
backend = AerSimulator(method="statevector")

qc_sdg = transpile(qc_sdg,backend)

result = backend.run(qc_sdg).result()

Statevector = result.get_statevector(qc_sdg, decimals=3)

print ("the quantum state is: ", Statevector)

plot_bloch_multivector(Statevector)
```

$\text{unitary}$

```python
#unitary matrix
qc_sdg = QuantumCircuit(1,1, name="qc")
qc_sdg.x(0)
qc_sdg.sdg(0)
backend = Aer.get_backend("unitary_simulator")

sdg_unitary = backend.run(qc_sdg)

sdg_unitary.result().get_unitary(qc_sdg)
```

$\text{T Gate}$

```python
qc_t = QuantumCircuit(1,1, name="qc")
qc_t.x(0)
qc_t.t(0)

qc_t.draw('mpl')
```

```python
backend = Aer.get_backend('statevector_simulator')

result = backend.run(qc_t).result().get_statevector(qc_t, decimals=3)

print("The quantum State is: ",result)

plot_bloch_multivector(result)
```

$\text{unitary}$

```python
#unitary

backend = Aer.get_backend('unitary_simulator')

t_unitary = backend.run(qc_t).result().get_unitary(qc_t, decimals=3)

print("Circuit unitary:\n", np.asarray(t_unitary).round(5))
```

$T^\dagger \text{ gate}$

```python
qc_tdg = QuantumCircuit(1,1, name="qc")
qc_tdg.x(0)
qc_tdg.tdg(0)

qc_tdg.draw('mpl')
```
