$\text{Constructing Classical Gate using Quatum Gates}$

$\text{Not Gate}$

```python
#Importing

import qiskit
import numpy as np
import math
from qiskit import QuantumCircuit,QuantumRegister,transpile
from qiskit.quantum_info import DensityMatrix, Operator
from qiskit_aer import AerSimulator, Aer
from qiskit.visualization import plot_histogram, plot_bloch_vector, plot_bloch_multivector, plot_state_city, plot_state_qsphere
from qiskit.quantum_info import Statevector
from  math import pi, sqrt

mpl = 'mpl'
```

```python
qc_not = QuantumCircuit(1,1, name='qc')
qc_not.x(0)
qc_not.measure([0],[0])
qc_not.draw(mpl)
```

$\text{QASM simulator}$

```python
#qasm simulator

backend = Aer.get_backend('qasm_simulator')

job = backend.run(qc_not, shots=1000)
result = job.result()

counts = result.get_counts(qc_not)

print("the total counts are: ", counts)

plot_histogram(counts)
```

$\text{AND gate}$

```python
qc_and = QuantumCircuit(3,1, name='qc')

qc_and.x(0)
qc_and.x(1)
qc_and.ccx(0,1,2)
qc_and.measure(2,0)

qc_and.draw(mpl)
```

$\text{QASM for AND Gate}$

```python
#qasm simulator

backend = Aer.get_backend('qasm_simulator')

job = backend.run(qc_and, shots=1000)
result = job.result()

counts = result.get_counts(qc_and)

print("the total counts are: ", counts)

plot_histogram(counts)
```

$\text{OR Gate}$

```python
qc_or = QuantumCircuit(3,1,name="qc")

qc_or.x(0)
qc_or.x(1)

qc_or.cx(1, 2)
qc_or.cx(0, 2)
qc_or.ccx(0, 1, 2)

qc_or.measure(2, 0)

qc_or.draw('mpl')
```

$\text{QASM simulator on OR Gate}$

```python
#qasm simulator

backend = Aer.get_backend('qasm_simulator')

circ = transpile(qc_or, backend)

job = backend.run(circ, shots=1000)



result = job.result()

counts = result.get_counts(qc_or)

print("the total counts are: ", counts)

plot_histogram(counts)
```

$\text{Haff Adder Circuit}$

```python
# Half Adder Circuit

qc_ha = QuantumCircuit(4,2,name="qc")

# Initialize qubits to |11>
qc_ha.x(0)
qc_ha.x(1)
qc_ha.barrier()

#XOR
qc_ha.cx(1, 2)
qc_ha.cx(0, 2)
qc_ha.barrier()

#AND
qc_ha.ccx(0, 1, 3)
qc_ha.barrier()

#Sum
qc_ha.measure(2, 0)

#Carry out
qc_ha.measure(3, 1)

qc_ha.draw('mpl')
```

$\text{QASM on Half Adder}$

```python
#qasm simulator

backend = Aer.get_backend('qasm_simulator')

circ = transpile(qc_ha, backend)

job = backend.run(circ, shots=1000)



result = job.result()

counts = result.get_counts(qc_ha)

print("the total counts are: ", counts)

plot_histogram(counts)
```
